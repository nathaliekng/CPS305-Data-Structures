## Solution for exercise 1

```python
import timeit
import matplotlib.pyplot as plt
import random

dic = {}
dic2 = {}
plt.xlabel('Size')
plt.ylabel('Time(in milliseconds)')
y1 = []
y2 = []
#x values are determined based on the range used to calculate the time
x1 = range(10000, 180000, 1000)

#First for loop meant to calculate time of get() and add values to y1 list for graph 
for i in range(10000, 180000, 1000):
    dic = {j: None for j in range(i)}
    tGet = timeit.timeit("dic.get(random.randrange(%d))"%i, "from __main__ import random, dic", number = 10)
    y1.append(tGet)
    #Same code to calculate time of update() and add values to y2 list
    dic2 = {j: None for j in range(i)}
    tUp = timeit.timeit("dic2.update({random.randrange(%d): ''})"%i, "from __main__ import random, dic2", number = 10)
    y2.append(tUp)

#plot values onto graph 
plt.plot(x1,y1,  "r+", label= "get()")
plt.plot(x1,y2, "b+", label = "update()")
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
plt.show()
```

**Discussion of the Results**
The graph comes out with a straight line (constant value) for both get and update methods. O(1) should show up as a constant straight line, therefore this proves that get() and update() are both O(1). The method get() shows to be slightly slower than update().

## Solution for Exercise 2

```python
import timeit
import matplotlib.pyplot as plt
import random

dic = {}
listy = []
y1 = []
y2 = []
x1 = range(10000, 190000, 1000)
plt.xlabel('Size')
plt.ylabel('Time(in milliseconds)')
plt.title('Running time of del operator between dictionary and list')

#dictionary is created and time is calculated for how long the delete operator takes for dictionary
#y values are appended to y1 list for graph
for i in range(10000, 190000, 1000):
    dic = {j: None for j in range(i)}
    tDic = timeit.timeit("del dic[random.randrange(%d)]"%i, "from __main__ import random, dic", number = 10)
    y1.append(tDic)

#list is created with the nested for loop and time is calculated the same way as it does for dictionary
#y values are appended to y2 list for graph
for i in range(10000, 190000, 1000):
    for y in range (i):
        listy.append(y)
    tList = timeit.timeit("del listy[random.randrange(%d)]"%i, "from __main__ import random, listy", number = 10)
    y2.append(tList)
    
#plot values onto graph
plt.plot(x1,y1,  "g+", label= "dictionary")
plt.plot(x1,y2, "r+", label = "list")
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
plt.show()
```

**Discussion of the Results**

We can see that the time it takes for the delete operator to work on list increases linearly, and the dictionary remains constant, and faster than list. The higher range we enter for list, the longer it will take and will continue to get significantly longer (in time).
